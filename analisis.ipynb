```python
from greedy_algorithm import obtener_orden_y_tiempo_optimo_de_videos
from random import randint
from time import time
import numpy as np
from matplotlib import pyplot as plt

def correr_benchmark():
    cantidades = [ i for i in range(1, 200000, 500) ]
    tiempos_algoritmo = []
    for cantidad in cantidades:
        tiempos = generar_tiempos_random(cantidad)
        empieza = time()
        obtener_orden_y_tiempo_optimo_de_videos(tiempos)
        termina = time()
        tiempos_algoritmo.append(termina - empieza)
    return cantidades, tiempos_algoritmo
        
def generar_tiempos_random(cantidad):
    return [(randint(1, 100), randint(1, 100)) for _ in range(cantidad)]

cantidades, tiempos = correr_benchmark()
plt.plot(np.array(cantidades), np.array(tiempos))
plt.show()
```
