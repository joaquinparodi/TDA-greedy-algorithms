# Analisis de tiempos del algoritmo

```python
from greedy_algorithm import obtener_orden_y_tiempo_optimo_de_videos
from random import randint
from time import time
import numpy as np
from matplotlib import pyplot as plt

def correr_benchmark(cantidad, max_tiempos):
    tiempos_video = generar_tiempos_random(cantidad, max_tiempos)
    
    empieza = time()
    obtener_orden_y_tiempo_optimo_de_videos(tiempos_video)
    termina = time()
    
    return termina - empieza
        
def generar_tiempos_random(cantidad, max):
    return [(randint(1, max), randint(1, max)) for _ in range(cantidad)]
```

### Variando el largo de la lista de videos

* En estas pruebas se vera como afecta la cantidad de videos al tiempo del algoritmo.
* Los Si y Ai (tiempos scaloni y tiempos ayudantes) seran random de 1 a 1000

```python

def plotear_tiempos_cantidad_variante():
    
    cantidades = [ i for i in range(0, 100001, 500) ]
    max_tiempos = 1000
    
    tiempos = []
    for index, cantidad in enumerate(cantidades):
        tiempos.append(correr_benchmark(cantidad, max_tiempos))
        if cantidad % 5000 == 0 and cantidad != 0 and cantidad <= 100000:
            plt.figure()
            plt.plot(np.array(cantidades[:index]), np.array(tiempos[:index]))
            plt.title("Benchmark con cantidad " + str(cantidad))
            plt.xlabel("Cantidad de videos (n)")
            plt.ylabel("Tiempo de ejecucion (s)")
            plt.show()

    plt.figure()
    plt.plot(np.array(cantidades), np.array(tiempos))
    plt.title("Benchmark con cantidad " + str(cantidades[-1]))
    plt.xlabel("Cantidad de videos (n)")
    plt.ylabel("Tiempo de ejecucion (s)")
    plt.show()

plotear_tiempos_cantidad_variante()
```

#### Se ve que a medida que la cantidad de videos aumenta el grafico del tiempo del algoritmo se asemeja a la funcion n*log(n)

### Variando Si y Ai (timpos scaloni y tiempos ayudantes)

* En estas pruebas se vera como afecta al tiempo del algoritmo cuando Si y Ai son numeros mas chicos o son numeros mas grandes.
* La cantidad de videos se fijara en 1000.

```python

def plotear_tiempos_duracion_variante():
    
    cantidad_videos = 1000
    max_tiempos = [ i for i in range(0, 10001, 10) ]
    max_tiempos[0] = 1
    
    tiempos = []
    for index, max_tiempo in enumerate(max_tiempos):
        tiempos.append(correr_benchmark(cantidad_videos, max_tiempo))
        if (max_tiempo % 1000 == 0 or max_tiempo == 100) and max_tiempo != 0:
            plt.figure()
            plt.plot(np.array(max_tiempos[:index]), np.array(tiempos[:index]))
            plt.title("Benchmark hasta tiempo maximo " + str(max_tiempo))
            plt.xlabel("Tiempo maximo")
            plt.ylabel("Tiempo de ejecucion (s)")
            plt.show()

    plt.figure()
    plt.plot(np.array(max_tiempos), np.array(tiempos))
    plt.title("Benchmark hasta tiempo maximo " + str(max_tiempos[-1]))
    plt.xlabel("Tiempo maximo")
    plt.ylabel("Tiempo de ejecucion (s)")
    plt.show()

plotear_tiempos_duracion_variante()
```

#### Se ve que el tiempo del algoritmo es independiente de la variabilidad de Si y Ai
