```python
from greedy_algorithm import obtener_orden_y_tiempo_optimo_de_videos
from random import randint
from time import time
import numpy as np
from matplotlib import pyplot as plt

def correr_benchmark(cantidad):
    tiempos_video = generar_tiempos_random(cantidad)
    
    empieza = time()
    obtener_orden_y_tiempo_optimo_de_videos(tiempos_video)
    termina = time()
    
    return termina - empieza
        
def generar_tiempos_random(cantidad):
    return [(randint(1, 100), randint(1, 100)) for _ in range(cantidad)]


def plotear_tiempos():
    
    cantidades = [ i for i in range(0, 200001, 500) ]
    tiempos = []
    for index, cantidad in enumerate(cantidades):
        tiempos.append(correr_benchmark(cantidad))
        if cantidad % 5000 == 0 and cantidad != 0 and cantidad <= 100000:
            plt.figure()
            plt.plot(np.array(cantidades[:index]), np.array(tiempos[:index]))
            plt.title("Benchmark con cantidad " + str(cantidad))
            plt.xlabel("Cantidad de videos")
            plt.ylabel("Tiempo de ejecucion")
            plt.show()

    plt.figure()
    plt.plot(np.array(cantidades), np.array(tiempos))
    plt.title("Benchmark con cantidad " + str(cantidades[-1]))
    plt.xlabel("Cantidad de videos")
    plt.ylabel("Tiempo de ejecucion")
    plt.show()

plotear_tiempos()
```
